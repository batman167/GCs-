1.functions without arguments and without return

```python
def one():
    print("hello")

one()
```

2.functions with arguments and without return

```python
def two(a,b):
    print(a+b)

two(1,2)
```

3.functions without arguments and with return

```python
def three():
    return None

print(three())
```

4.functions with arguments and with return

```python
def add(a,b):
    return a+b

print(add(1,2))
```
