```python
import random
import numpy as np
import pandas as pd

data_ranges = {
    "age": (50, 80),
    "age_normalized": (1, 4),
    "sex": (0, 1),
    "smoking": (0, 2),
    "survival": (0, 1),
    "sur_time_mon": (1, 50),
    "pro_time_mon": (1, 30),
    "VDW": (-55, -45),
    "EEL": (-30, -10),
    "EGB": (30, 45),
    "EEL.1": (-30, -10),
    "EPB": (35, 50),
    "ENPOLAR": (-40, -30),
    "TOTAL.1": (-80, -50),
    "matching_rates": (0.3, 0.5),
    "centroid_distance": (45, 50),
    "connectivity": (15, 25),
    "convex_atoms": (20, 30),
    "hydrogen_bonds": (900, 1700),
    "response": (0, 1)
}

# Generating 1000 rows of synthetic data
num_rows = 100
synthetic_data = {key: [random.uniform(*data_ranges[key]) if isinstance(data_ranges[key][0], float) else random.randint(*data_ranges[key]) for _ in range(num_rows)] for key in data_ranges}

# Converting the synthetic data into a dataframe
synthetic_df = pd.DataFrame(synthetic_data)

synthetic_df
```
