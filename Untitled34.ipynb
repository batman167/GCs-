```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder

# Load your data into a DataFrame
data = pd.read_csv('sample_data.csv')

# Separate features and target
X = data.drop('response', axis=1)
y = data['response']

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the target variable
y = label_encoder.fit_transform(y)

# Apply SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# Combine into a new DataFrame
augmented_data = pd.DataFrame(X_res, columns=X.columns)

# Inverse transform the encoded target variable to original labels
augmented_data['response'] = label_encoder.inverse_transform(y_res)

# Print class counts after SMOTE
print(augmented_data['response'].value_counts())
```

```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer  # Import SimpleImputer

# Load your data into a DataFrame
data = pd.read_csv('sample_data.csv')

# Separate features and target
X = data.drop('response', axis=1)
y = data['response']

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the target variable
y = label_encoder.fit_transform(y)

# Impute missing values using SimpleImputer # New lines to handle NaN
imputer = SimpleImputer(strategy='mean') # You can choose a different strategy if needed
X = imputer.fit_transform(X)

# Apply SMOTE with a reduced number of neighbors
smote = SMOTE(random_state=42, k_neighbors=1) # Setting k_neighbors to 2
X_res, y_res = smote.fit_resample(X, y)

# Combine into a new DataFrame
augmented_data = pd.DataFrame(X_res, columns=X.columns)

# Inverse transform the encoded target variable to original labels
augmented_data['response'] = label_encoder.inverse_transform(y_res)

# Print class counts after SMOTE
print(augmented_data['response'].value_counts())
```
