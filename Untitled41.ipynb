```python
import pymol
from pymol import cmd

# Load the PDB structure and highlight the mutation site
pymol.finish_launching()

# Load the provided PDB file
pdb_file_path = '2ity.cif.pdb'
cmd.load(pdb_file_path)

# Select and highlight residue 858 in chain A (LEU mutation)
cmd.select("mutation_site", "chain A and resi 858")
cmd.show("sticks", "mutation_site")
cmd.color("red", "mutation_site")

# Zoom into the highlighted mutation site
cmd.zoom("mutation_site")

# Save the image
output_image_path = '/mnt/data/highlighted_mutation_site.png'
cmd.png(output_image_path)

output_image_path
```
