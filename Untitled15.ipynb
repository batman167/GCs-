```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
np.random.seed(42)
n_samples = 20000

data = {
    'Age': np.random.uniform(30, 80, n_samples),
    'Age_normalized': np.random.randint(1, 6, n_samples),
    'Sex': np.random.randint(0, 1, n_samples),
    'Smoking': np.random.randint(0, 2, n_samples),
    'Tumor_Size': np.random.uniform(0.5, 10, n_samples),
    'Tumor_Grade': np.random.randint(1, 4, n_samples),
    'ER_Status': np.random.choice(['Positive', 'Negative'], n_samples),
    'PR_Status': np.random.choice(['Positive', 'Negative'], n_samples),
    'HER2_Status': np.random.choice(['Positive', 'Negative'], n_samples),
    'Lymph_Node_Involvement': np.random.randint(0, 31, n_samples),
    'VDW': np.random.uniform(-60, -40, n_samples),
    'EEL': np.random.uniform(-30, -10, n_samples),
    'EPB': np.random.uniform(30, 50, n_samples),
    'ENPOLAR': np.random.uniform(-60, -40, n_samples),
    'TOTAL': np.random.uniform(-80, -30, n_samples),
    'Matching_rates': np.random.uniform(0.2, 0.8, n_samples),
    'Centroid_distance': np.random.uniform(20, 50, n_samples),
    'Connectivity': np.random.randint(10, 31, n_samples),
    'Convex_atoms': np.random.randint(4, 16, n_samples),
    'Hydrogen_bonds': np.random.randint(2, 11, n_samples),
    'Response': np.random.choice(['Stable', 'No'], n_samples),
}

df = pd.DataFrame(data)
df.to_csv('BC_dataset.csv', index=False)
```

```python
sum(df['Sex'] == 0)
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
lr = LogisticRegression()
rf = RandomForestClassifier()
svm = SVC()
mlp = MLPClassifier()
xb = xgb.XGBClassifier()
ss = StandardScaler()
le = LabelEncoder()
```

```python
df['Response'] = le.fit_transform(df['Response'])
df['ER_Status'] = le.fit_transform(df['ER_Status'])
df['PR_Status'] = le.fit_transform(df['PR_Status'])
df['HER2_Status'] = le.fit_transform(df['HER2_Status'])
```

```python
df
```

```python
X = df.drop(['Response'], axis=1)
y = df['Response']

X_scaled = ss.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
lr.fit(X_train, y_train)
rf.fit(X_train, y_train)
svm.fit(X_train, y_train)
mlp.fit(X_train, y_train)
xb.fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
y_pred_rf = rf.predict(X_test)
y_pred_svm = svm.predict(X_test)
y_pred_mlp = mlp.predict(X_test)
y_pred_xb = xb.predict(X_test)
```

```python
print(f"Logistic Regression Accuracy: {accuracy_score(y_test, y_pred_lr)}")
print(f"Random Forest Accuracy: {accuracy_score(y_test, y_pred_rf)}")
print(f"SVM Accuracy: {accuracy_score(y_test, y_pred_svm)}")
print(f"MLP Accuracy: {accuracy_score(y_test, y_pred_mlp)}")
print(f"XGBoost Accuracy: {accuracy_score(y_test, y_pred_xb)}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)

print(f"KNN Accuracy: {accuracy_score(y_test, y_pred_knn)}")
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.heatmap(confusion_matrix(y_test, y_pred_xb), annot=True, fmt="d", cmap="Blues", cbar=False)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```
