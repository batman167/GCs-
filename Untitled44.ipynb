```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

data = pd.read_csv('/content/EGFR_3 (modified).csv')

# Preprocessing the target variable
le = LabelEncoder()
data['response'] = le.fit_transform(data['response'])

# Separating features and target
X = data.drop(columns=['response'])
y = data['response']

# Scaling features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Define the DNN model
model = Sequential([
    Dense(64, input_dim=X_train.shape[1], activation='relu'),  # Input and first hidden layer
    Dropout(0.3),  # Regularization
    Dense(32, activation='relu'),  # Second hidden layer
    Dropout(0.3),  # Regularization
    Dense(len(np.unique(y)), activation='softmax')  # Output layer for multi-class classification
])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Fit the model
history = model.fit(X_train, y_train, validation_split=0.2, epochs=100, batch_size=8, verbose=1)

# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)

test_loss, test_accuracy
```

```python
from tensorflow.keras.layers import Input, Dense, LayerNormalization, MultiHeadAttention, Dropout
from tensorflow.keras.models import Model
```

```python
X = data.drop(columns=['response'])
y = data['response']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Apply SMOTE
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Reshape for Transformer input
X_train_smote = tf.expand_dims(X_train_smote, axis=-1)
X_test = tf.expand_dims(X_test, axis=-1)

# Transformer block
def transformer_block(inputs, head_size, num_heads, ff_dim, dropout=0.3):
    attention = MultiHeadAttention(num_heads=num_heads, key_dim=head_size)(inputs, inputs)
    attention = Dropout(dropout)(attention)
    attention = LayerNormalization(epsilon=1e-6)(attention + inputs)

    ff = Dense(ff_dim, activation="relu")(attention)
    ff = Dropout(dropout)(ff)
    ff = Dense(inputs.shape[-1])(ff)
    return LayerNormalization(epsilon=1e-6)(ff + attention)

# Build model
input_layer = Input(shape=(X_train_smote.shape[1], X_train_smote.shape[2]))
x = transformer_block(input_layer, head_size=32, num_heads=4, ff_dim=64)
x = Dense(32, activation="relu")(x)
x = Dropout(0.3)(x)
x = tf.keras.layers.GlobalAveragePooling1D()(x)
output_layer = Dense(len(le.classes_), activation="softmax")(x)

model = Model(inputs=input_layer, outputs=output_layer)
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

# Train the model
history = model.fit(X_train_smote, y_train_smote, validation_split=0.2, epochs=100, batch_size=16, verbose=1)

# Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Loss: {test_loss:.4f}, Test Accuracy: {test_accuracy:.4f}")
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Embedding, Dense, Dropout, LayerNormalization, Attention
from tensorflow.keras.models import Model
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import numpy as np
```

```python
le = LabelEncoder()
data['response'] = le.fit_transform(data['response'])

# Separate features and target
X = data.drop(columns=['response'])
y = data['response']

# Identify categorical and numerical features
categorical_columns = []  # Example categorical columns
numerical_columns = [col for col in X.columns if col not in categorical_columns]

# Standardize numerical features
scaler = StandardScaler()
X[numerical_columns] = scaler.fit_transform(X[numerical_columns])

# Encode categorical features
X_encoded = X.copy()
for col in categorical_columns:
    X_encoded[col] = LabelEncoder().fit_transform(X[col])

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42, stratify=y)

# Apply SMOTE to address class imbalance
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Convert to numpy arrays
X_train_smote = np.array(X_train_smote)
X_test = np.array(X_test)

# Reshape for RNN/TabTransformer input
X_train_smote_rnn = X_train_smote.reshape((X_train_smote.shape[0], X_train_smote.shape[1], 1))
X_test_rnn = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# TabTransformer Model Definition
embedding_size = 8  # Define embedding size for categorical columns

inputs = []
embeddings = []

# Categorical embeddings
for col in categorical_columns:
    input_layer = Input(shape=(1,))
    embedding_layer = Embedding(input_dim=len(np.unique(X[col])), output_dim=embedding_size)(input_layer)
    inputs.append(input_layer)
    embeddings.append(embedding_layer)

# Concatenate numerical features
numerical_input = Input(shape=(len(numerical_columns),))
inputs.append(numerical_input)

# Combine embeddings and numerical features
x = tf.keras.layers.Concatenate()(embeddings + [numerical_input])

# Transformer Block (Self-Attention)
x = LayerNormalization()(x)
x = Attention()([x, x])  # Self-attention mechanism
x = Dropout(0.3)(x)

# Output Layer
x = Dense(64, activation='relu')(x)
x = Dropout(0.3)(x)
output = Dense(1, activation='sigmoid')(x)  # For binary classification

# Create the model
model = Model(inputs=inputs, outputs=output)

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit([X_train_smote[col].values for col in categorical_columns] + [X_train_smote[numerical_columns].values],
                    y_train_smote, validation_data=([X_test[col].values for col in categorical_columns] + [X_test[numerical_columns].values], y_test),
                    epochs=50, batch_size=32, verbose=1)

# Evaluate the model
test_loss, test_accuracy = model.evaluate([X_test[col].values for col in categorical_columns] + [X_test[numerical_columns].values], y_test)
test_loss, test_accuracy
```
