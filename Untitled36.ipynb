```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('EGFR_3.csv')
```

```python
df['response'].unique()
```

```python
import xgboost as xgb
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, roc_auc_score, auc, f1_score, precision_score, recall_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
le = LabelEncoder()
df['response'] = le.fit_transform(df['response'])
```

```python
X = df.drop('response', axis=1)
y = df['response']

X = StandardScaler().fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)
```

```python
model = xgb.XGBClassifier()
model.fit(X_train, y_train)
```

```python
y_prediction = model.predict(X_test)
```

```python
accuracy_score(y_test, y_prediction)
```

```python
print(classification_report(y_test, y_prediction))
```

```python
confusion_matrix(y_test, y_prediction)
```

```python
recall_score(y_test, y_prediction)
```

```python
sns.heatmap(confusion_matrix(y_test, y_prediction), annot=True, fmt='d', cmap='Blues')
```
